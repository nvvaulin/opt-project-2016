The task is to make a proper schedule for Skoltech.

The schedule must define for each subject: auditorium, time slots, days of the week. One course can not take place several times during one day, with the exception for "Basic Molecular Biology Techniques+Seminar", it takes the all Wednesday.

The constraints are given in 4 tables.

<ol>
 <li>'CoursesOverlapping': 
<p>For each pair of subjects we know whether they can overlap or not (the same day, the same time). </p></li>
 <li> 'CoursesWorkload': 
<p>Number of session per week and the session length is given.</p></li>
<li>'Courses vs Time-slots' :
<p>Preferable time slots for each course:
       <ul>
        <li>-1 - the time slot is totally inapplicable for the course because of the course type (see session lengths);</li>
        <li>0 - the time slot is very inconvenient;</li>
        <li>1 - the time slot is okay, but could be better;</li>
        <li>2 - the time slot can be used for the course session;</li>
        </ul>
</p>         
       
<li>'Courses vs Classrooms': 
<p>Preferable auditoriums for different courses: 
       <ul>
        <li>-1 - the auditorium is totally inapplicable for the course ;</li>
        <li>0 - the auditorium is very inconvenient; </li>
        <li>1 - the auditorium is okay, but could be better;</li>
        <li>2 - the auditorium can be used for the course session;</li>
       </ul>        
</p>
</li> 
</ol>
Also, take into account weekly events : 
Energy Colloquium and Skoltech Colloquium, Tue (4-5pm) and Thu (4-5pm) respectively, both are held in room 402.

```python
def row_col_pref(TT):
    TT_res = np.reshape(TT, (clas*Time,1))
    max_num = np.argmax(TT_res)
    max_el = np.max(TT_res)
    col = max_num - clas*np.round(max_num/clas)
    row = np.round(max_num/clas)
    return row, col, max_el

def change_cond(a3, a4):
    for i  in range(subj):
        for j in range(clas):
            if a4[j, i] == -1:
                a4[j, i] = -100
    for i  in range(subj):
        for j in range(Time):
            if a3[j,i]== -1:
                a3[j,i] = -100
    return a3, a4
```

```python
def intersection(list1, list2):
    inter_list = []
    for i in list1:
        for j in list2:
            if i == j:
                inter_list.append(i)
                break
    return inter_list
```

```python
def double_time(Time):
    if Time ==0:
        dtime = [3,4]
    if Time ==1:
        dtime = [4,5]
    if Time ==2:
        dtime = 6
    if Time ==3:
        dtime = 0
    if Time ==4:
        dtime = [0, 1]
    if Time ==5:
        dtime = 1
    if Time == 6:
        dtime = 2
    return dtime



def get_time_clas_table(a33, a44):
    for i in range(0,subj):
        for j in range(clas):
            for l in range(Time):
                T[i,l,j] = a33[l,i+1] + a44[j,i+1]
    return T

```

```python
def row_col_pref(TT):
    TT_res = np.reshape(TT, (clas*Time,1))
    max_num = np.argmax(TT_res)
    max_el = np.max(TT_res)
    col = max_num - clas*np.round(max_num/clas)
    row = np.round(max_num/clas)
    return row, col, max_el
```

```python

def get_TimeTable(a3, a4, A):
    TimeTable =  np.chararray((days,Time,sub+1), itemsize=30)
    TimeTable[:] = ' '
    for d in range(days):
        for i in range(Time):
            for j in range(sub):
                for k in range(clas):
                    if A[j, d, i, k] > 0:
                        TimeTable[d, i, j+1] = a4[k,0]
                    #else:
                     #   TimeTable[d, i, j+1] = ' '
        TimeTable[d, :,0] = a3[:,0]
        #TimeTable[d, 0, sub] = 'Energy Colloquium'
        #TimeTable[d, 0, sub - 1] = 'Skoltech Colloquium'
        #TimeTable[d, 0, :sub-2] = a3[0,1:24]
    return TimeTable
```

```python
clas = 9
sub =  25
Time = 7
days = 5
global clas, sub, Time, days
A = np.zeros((sub, days, Time, clas))
A[0, 2, :3, 8] = 1 #fix Biology
A[23, 1, 6,0 ] = 1 # fix colloquims
A[24, 3, 6,0] = 1
subj = sub -2
global sub

ListSub = []
for i in range(sub):
    ListSub.append([])

for i in range(sub):
    for j in range(sub):
        if a1[i,j] == 0:
            ListSub[i].append(j)

a33, a44 =  change_cond(a3,a4)
T = np.zeros((subj, Time, clas))
T = get_time_clas_table(a33, a44)
```

```python
a2_chan = np.copy(a2)
a3_chan, a4_chan =  change_cond(a3,a4)
T = get_time_clas_table(a3_chan, a4_chan)
import copy

a2_chan[0,:] = 0;
rest_list = range(1,subj)
sum_pref = np.zeros(subj)
sum_pref[0] = 1000
# continue fill in table
for j in range(days):# days
    T_day = np.copy(T)
    ListTime = []
    for i in range(Time):
        ListTime.append([])
    if j ==2:# fix biology
        ListTime[0].append(0)
        ListTime[1].append(0)
        ListTime[2].append(0)
    if j==1:#fix colloquiums
        ListTime[6].append(24)
    if j==3:
        ListTime[6].append(23)
    list_subj = copy.copy(rest_list)
    time_classes_table = np.zeros((Time,clas))
    if j ==2:# fix biology
        time_classes_table[0, 8]
        time_classes_table[1, 8]
        time_classes_table[2, 8]
    if j==1:#fix colloquiums
        time_classes_table[6,0]
        time_classes_table[2,0]
    if j==3:
        time_classes_table[6, 0]
        time_classes_table[2, 0]
    
    ind2 = True
    while (np.sum(time_classes_table) != Time*clas):
        for i in list_subj:#subject
            sum_pref[i] = np.sum(a4[:,i+1]) ###+np.sum(a3[:,i+1])
        min_pref_sub = np.argmin(sum_pref)
        if min_pref_sub == 0:
            break
        ind = True
        while ind == True:
            #time_pref = np.argmax(a3[:,min_pref_sub+1])
            #clas_pref = np.argmax(a3[:,min_pref_sub+1])
            time_pref, clas_pref, max_el = row_col_pref(T_day[min_pref_sub,:,: ])
            if max_el<0:
                ind2 = False
                break
            if time_classes_table[time_pref, clas_pref] == 0 and intersection(ListTime[time_pref], ListSub[min_pref_sub])==[]:
                A[min_pref_sub, j, time_pref, clas_pref] = 1
                ListTime[time_pref].append(min_pref_sub)
                time_classes_table[time_pref, clas_pref] = 1
                dtime = double_time(time_pref)
                time_classes_table[dtime, clas_pref] = 1
                ind = False
            else:
                T_day[min_pref_sub,time_pref,clas_pref] = -100
                
                                   
        if ind2 == False:
            break
        a2_chan[min_pref_sub,0] = a2_chan[min_pref_sub,0] - 1
        
        if  a2_chan[min_pref_sub,0] == 0:
            rest_list.remove(min_pref_sub)
        sum_pref[min_pref_sub] = 1000
        list_subj.remove(min_pref_sub) 
```
